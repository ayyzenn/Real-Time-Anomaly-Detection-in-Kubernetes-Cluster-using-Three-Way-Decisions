```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statsmodels.nonparametric.smoothers_lowess import lowess
from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.seasonal import MSTL

sns.set_context("talk")

#Load Data
pd.set_option('display.max_rows', None)
df = pd.read_csv("./dataset/A4Benchmark/A4Benchmark-TS1.csv")

#print data
df['timestamps'] = pd.to_datetime(df['timestamps'], unit='s')
df
```

```python
# Plot data to visually inspect any outliers
fig, ax = plt.subplots(figsize=[10, 5])
df.plot(y="value", marker=".", figsize=[14, 6], legend=None, ax=ax)
ax.set_title("Visually Inspected outliers")
ax.set_ylabel("Values")
ax.set_xlabel("Time")
```

```python
# Deseasonalised data using MSTL
mstl = MSTL(df["value"], periods=(12, 23, 24, 168), stl_kwargs={"seasonal_deg": 0})
#mstl = MSTL(df["value"], periods=(12, 23, 24, 48, 168), stl_kwargs={"seasonal_deg": 0})
res = mstl.fit()
seasonal_component = res.seasonal
seasonal_component

s_sum = seasonal_component['seasonal_12']+seasonal_component['seasonal_23']+seasonal_component['seasonal_24']+seasonal_component['seasonal_168']
#s_sum.head()

#s_sum = seasonal_component['seasonal_12']+seasonal_component['seasonal_24']+seasonal_component['seasonal_48']+seasonal_component['seasonal_168']

df_deseasoned = (df["value"] - s_sum).to_frame("value")

# Plot Deseasonalised data
fig, ax = plt.subplots(figsize=[14, 6])
df_deseasoned.plot(y="value", marker=".", figsize=[14, 6], legend=None, ax=ax)
ax.set_title("Deseasonalised Data with outliers")
ax.set_ylabel("Deseasonalised Values")
ax.set_xlabel("Time")
```

# Fit a trend line using LOWESS and examine residuals

```python
df_ = df_deseasoned.copy()
```

```python
y = df_["value"].values
# The x-values must be numbers rather than dates to fit the LOWESS model.
x = np.arange(0, len(y), 1)
trend = lowess(y, x, frac=0.40)
df_["trend"] = trend[:, 1]

ax = df_.plot(y=["value", "trend"], marker=".", figsize=[14, 6])
ax.set_title("Deseasonalised Data with outliers")
ax.set_ylabel("Deseasonalised Values")
ax.set_xlabel("Time")
```

```python
# Compute residuals
df_["residual"] = df_["value"] - df_["trend"]

# Compute inter-quartile range
Q1 = df_["residual"].quantile(0.25)
Q3 = df_["residual"].quantile(0.75)
IQR = Q3 - Q1

#___________________________________________________
# df_["q1"] = Q1
# df_["q3"] = Q3
#___________________________________________________

# Apply 1.5 IQR as a threshold to identify outliers
factor = 1.5
df_["lower"] = Q1 - factor * IQR
df_["upper"] = Q3 + factor * IQR
df_["is_outlier"] = (df_["residual"] < df_["lower"]) | (
    df_["residual"] > df_["upper"]
)

df_
```

# Let's plot the results

```python
# Plot time series
fig, ax = plt.subplots(nrows=2, figsize=[12, 10], sharex=True)
df_.plot(y=["value", "trend"], marker=".", ax=ax[0])
df_["residual"] = df_["residual"]

# Plot residuals
df_.plot(y=["residual"], ax=ax[1], marker="o", linestyle="")
df_.plot(y=["upper", "lower"], ax=ax[1], color="k", legend=None, alpha=0.9)

# If any data points are identified as an outlier, plot them
if df_["is_outlier"].any():
    df_["value"].loc[df_["is_outlier"]].plot(
        marker="o", color="r", ax=ax[0], legend=None, linestyle="", alpha=0.5
    )
    df_["residual"].loc[df_["is_outlier"]].plot(
        marker="o", color="r", ax=ax[1], legend=None, linestyle=""
    )

ax[0].set_title("Deseasonalised Data with outliers")
ax[0].set_ylabel("Deseasonalised Values")
ax[1].set_ylabel("Residuals")
ax[1].set_xlabel("Time")
plt.tight_layout()
```

# We can also visualise the residuals for outliers using a histogram and box plot. This also allows us to check that the data is approximately normal.

```python
# ACF
fig, ax = plt.subplots(figsize=[12, 6])
plot_acf(
    x=df_['residual'],
    lags=100,
    ax=ax,
    alpha=0.05,
    auto_ylims=True
)
ax.set_title("Autocorrelation of residual")
ax.set_ylabel("Autcorrelation")
ax.set_xlabel("Lag")
plt.tight_layout()
```

```python
f, (ax_box, ax_hist) = plt.subplots(
    2, sharex=True, gridspec_kw={"height_ratios": (0.20, 0.80)}, figsize=[12, 7]
)
sns.boxplot(x=df_["residual"], ax=ax_box)
sns.histplot(x=df_["residual"], ax=ax_hist)

ax_box.axes.get_xaxis().set_visible(False)
ax_box.axes.get_yaxis().set_visible(False)
ax_box.set_title('Box plot and histogram of residuals')
    
ax_hist.vlines(Q1 - factor * IQR, 0, 40, color="r")
ax_hist.vlines(Q3 + factor * IQR, 0, 40, color="r")
sns.despine()
plt.show()
```

```python
import numpy as np

def calculate_rmse(actual_values, predicted_values):

    # Convert inputs to numpy arrays
    actual_values = np.array(actual_values)
    predicted_values = np.array(predicted_values)
    
    # Calculate the squared differences
    squared_diff = (actual_values - predicted_values) ** 2
    
    # Calculate the mean squared error
    mean_squared_error = np.mean(squared_diff)
    
    # Calculate the root mean squared error
    rmse = np.sqrt(mean_squared_error)
    
    return rmse

# Example usage:
actual_values = df["trend"]
predicted_values = df_["trend"]

rmse_value = calculate_rmse(actual_values, predicted_values)
print("Root Mean Square Error (RMSE):", rmse_value)
```
