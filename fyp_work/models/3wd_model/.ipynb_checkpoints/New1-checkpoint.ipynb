# Explanation of Sliding Window Variance Calculation

### Initializing Variables
An empty dictionary named `window_variances` is initialized. This dictionary will be used to store the calculated variances for each window.

### Iterating over the DataFrame with a Fixed Step Size
A `for` loop iterates over the DataFrame `df` with a fixed step size. The loop ranges from index 0 to the length of `df` minus the step size plus 1, with each iteration incrementing by the step size.

### Selecting the Window of Data
Within each iteration, a window of data is selected using the `iloc` method to slice the DataFrame `df` from index `i` to `i + step_size`. This window represents the subset of data used for variance calculation.

### Calculating the Mean of the Window Data
The mean of the values within the selected window is calculated using the `mean()` method of the `'value'` column. It is assumed that `'value'` is the column containing the time series data.

### Calculating the Squared Differences from the Mean
The squared differences between each value in the window and the mean of the window are computed. This is achieved by subtracting the window mean from each value and squaring the result.

### Calculating the Variance of the Window Data
The variance of the window data is calculated manually using the formula for variance: the sum of squared differences divided by the degrees of freedom, which is `(step_size - 1)`.

### Storing the Variance in the Dictionary
The calculated variance is stored in the `window_variances` dictionary using a key that represents the index of the current window. The key format is `'window#N'`, where `N` represents the index of the window.

### Printing the Dictionary of Window Variances
After iterating through all windows, the dictionary `window_variances` containing the variances for each window is printed.



```python
import pandas as pd

# Load the dataset
df = pd.read_csv("./dataset/A4Benchmark/test.csv")
# df = pd.read_csv("./dataset/A4Benchmark/A4Benchmark-TS1.csv")

# Define the fixed step size
step_size = 5 #window size

# Initialize an empty dictionary to store window variances
window_variances = {}

# Iterate over the DataFrame with a fixed step size
for i in range(0, len(df) - step_size + 1, step_size):
    # Select the window of data
    window_data = df.iloc[i:i + step_size]
    
    # Calculate the mean of the window data
    window_mean = window_data['value'].mean()  # Assuming 'value' is the column containing your time series data
    # print(window_mean,'\n')
    
    # Calculate the squared differences from the mean
    sub = window_data['value'] - window_mean
    # print("Sub is:",sub, '\n+++++ \n')
    squared_diffs = (sub) ** 2
    # print("Sq is:",squared_diffs, '\n+++++ \n')
    
    # Calculate the variance of the window data (manually)
    variance = squared_diffs.sum() / (step_size - 1)
    
    # Store the variance in the dictionary
    window_variances[f'window#{i//step_size + 1}'] = variance

# Print the dictionary of window variances
print(window_variances)
```

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
df = pd.read_csv("./dataset/A4Benchmark/test.csv")

# Define the window size
window_size = 5

# Initialize an empty dictionary to store window standard deviations
window_std_devs = {}

# Iterate over the DataFrame with a sliding window
for i in range(len(df) - window_size + 1):
    # Select the window of data
    window_data = df.iloc[i:i + window_size]

    print(window_data.head(),'\n')
    
    # Calculate the mean of the window data
    window_mean = window_data['value'].mean()  # Assuming 'value' is the column containing your time series data
    
    # Calculate the squared differences from the mean
    squared_diffs = (window_data['value'] - window_mean) ** 2
    
    # Calculate the variance of the window data
    variance = squared_diffs.sum() / (window_size - 1)
    
    # Calculate the standard deviation as the square root of variance
    std_dev = np.sqrt(variance)
    
    # Store the standard deviation in the dictionary
    window_std_devs[f'window#{i+1}'] = std_dev

# Print the dictionary of window standard deviations
print(window_std_devs)
```

# standard deviation

```python
def sliding_window_std(lst, window_size):
    results = []
    
    for i, num in enumerate(lst):
        # Calculate the window boundaries
        start_index = max(0, i - window_size // 2)
        end_index = min(len(lst), i + window_size // 2 + 1)
        
        # Get the numbers within the window
        window = lst[start_index:end_index]
        
        # Calculate the average of the window
        avg = sum(window) / len(window)
        
        # Calculate the squared differences from the average
        squared_diffs = [(avg - n) ** 2 for n in window]
        
        # Calculate the sum of squared differences
        sum_squared_diffs = sum(squared_diffs)
        
        # Calculate the standard deviation of the window
        std_dev = (sum_squared_diffs / len(window)) ** 0.5
        
        # Append the result to the results list
        results.append((num, window, std_dev))
        
    return results

lst = [509.276632, 557.008664, 619.458683, 623.929433, 633.684138, 653.559915, 566.842374, 774.303327, 531.359225, 452.040120,
       472.528354, 491.361270, 539.037251, 591.125394]
window_size = 5

std_devs = sliding_window_std(lst, window_size)
for num, window, std_dev in std_devs:
    print(f"{num}, {window} -> {std_dev}")
```

```python
def sliding_window_avg(lst, window_size, threshold):
    results = []
    stored_values = []

    for i, num in enumerate(lst):
        if i < window_size // 2:
            window = lst[:i + window_size // 2 + 1]
        elif i >= len(lst) - window_size // 2:
            window = lst[i - window_size // 2:]
        else:
            window = lst[i - window_size // 2:i + window_size // 2 + 1]
    
        avg = sum(window) / len(window)
        
        diff_sum = sum(abs(avg - n) for n in window)
        
        result = diff_sum / len(window)
        results.append(result)

        if len(results) >= 2:
            avg_prev_results = sum(results[-2:]) / 2
            threshold_value = avg_prev_results + threshold

            if result > threshold_value:
                stored_values.append((i - window_size // 2, window[-1]))

    return stored_values

# lst = [4, 5, 6, 7, 9, 10, 8, 120, 14, 15, 17, 2, 19, 20, 21, 18, 17]
lst = [509.276632, 557.008664, 619.458683, 623.929433, 633.684138, 653.559915, 566.842374, 774.303327, 531.359225, 
       452.040120, 472.528354, 491.361270, 539.037251, 591.125394]

window_size = 5
threshold = 5

stored_values = sliding_window_avg(lst, window_size, threshold)

print("Stored Values:")
for index, value in stored_values:
    print(f"Index: {index}, Value: {value}")
```
