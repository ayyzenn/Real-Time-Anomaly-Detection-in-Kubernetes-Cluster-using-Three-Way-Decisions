```python
import datetime
import requests
import json
from time import sleep
import csv
import threading
import pandas as pd
import time
from scipy.stats import median_abs_deviation
import os


# Prometheus API endpoint
prometheus_api = "http://192.168.49.2:31090/api/v1/query"

# Prometheus query
query = "sum(rate(container_cpu_usage_seconds_total{pod=~\"front-end-.*\", namespace=\"sock-shop\"}[1m])) by (pod_name)"

# CSV file path
csv_file_path = "cpu_data_frontend.csv"

# CSV file header
csv_header = ["Timestamp", "Value"]

# Lock for synchronization
lock = threading.Lock()

def fetch_data():
    while True:
        # Fetch data from Prometheus
        params = {"query": query}
        response = requests.get(prometheus_api, params=params)
        data = response.json()

        # Open CSV file in append mode
        with open(csv_file_path, mode="a", newline="") as csv_file:
            writer = csv.writer(csv_file)

            # Write header if file is empty
            if csv_file.tell() == 0:
                writer.writerow(csv_header)

            # Write data to CSV
            for result in data["data"]["result"]:
                time = result["value"][0]
                dt_object = datetime.datetime.fromtimestamp(time)
                timestamp = dt_object.strftime('%Y-%m-%d %H:%M:%S')
                value = float(result["value"][1])
                writer.writerow([timestamp, value])
        
        # Sleep for 5 seconds
        sleep(5)

def detect_anomaly():
    while True:
        
        while not os.path.exists(csv_file_path) or get_csv_row_count() < 5:
            sleep(5)

         # Read CSV file
        data = pd.read_csv(csv_file_path)
        
        new_data = data[-5:]
        data1 = detect_anomaly_inner(new_data)
        print()

        # Drop unnecessary columns
        column_to_drop = 'z-score'
        data1 = data1.drop(column_to_drop, axis=1)


        df = pd.read_csv('frontend_cpu_result.csv')
        if df.empty:
             data1.to_csv('frontend_cpu_result.csv', mode='a', index=False, header=False)
        else:
            last_value = df.iloc[-1]['Time']
            time2_dt = datetime.datetime.strptime(last_value, '%Y-%m-%d %H:%M:%S')
            for value in data1["Timestamp"]:
                time1_dt = datetime.datetime.strptime(value, '%Y-%m-%d %H:%M:%S')
                if time1_dt > time2_dt:
                    matching_rows = data1[data1["Timestamp"] == value]  
                    matching_rows.to_csv('frontend_cpu_result.csv',mode='a', header=False,  index=False)
                else:
                    pass



        

    # Sleep for 5 seconds
    time.sleep(5)

def detect_anomaly_inner(data):
    data['z-score'] = data['Value'].apply(compute_robust_z_score)
    data['is_anomaly'] = 1
    data.loc[data['z-score'] >= 3.5, 'is_anomaly'] = -1
    data.loc[data['z-score'] <= -3.5, 'is_anomaly'] = -1
    return data

def compute_robust_z_score(x):
    return .6745 * (x - median) / mad

def get_csv_row_count():
    if os.path.exists(csv_file_path):
        with open(csv_file_path, mode="r") as csv_file:
            reader = csv.reader(csv_file)
            row_count = sum(1 for _ in reader)
        return row_count
    else:
        return 0


def main():
    # Start fetch_data thread
    fetch_data_thread = threading.Thread(target=fetch_data)
    fetch_data_thread.start()

    # Start detect_anomaly thread
    detect_anomaly_thread = threading.Thread(target=detect_anomaly)
    detect_anomaly_thread.start()

if __name__ == '__main__':
    
    main()
    
```
