```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Directory where your CSV files are located
csv_directory = './A1Benchmark/'

# Desired column name to count ones
desired_column = 'is_anomaly'

# Initialize an empty dictionary to store the results
result_dict = {}

# Loop through each CSV file in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_directory, filename)

        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Count the number of ones in the desired column
        ones_count = df[desired_column].sum()

        # Store the result in the dictionary with the file name as the key
        result_dict[filename] = ones_count

# Print the dictionary with the results
print(result_dict)

# Plot the results
file_names = list(result_dict.keys())
one_counts = list(result_dict.values())

plt.figure(figsize=(10, 6))
plt.bar(file_names, one_counts)
plt.xlabel('CSV Files')
plt.ylabel('Count of Ones')
plt.title('Count of Ones in CSV Files')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```
